```python
#separating the compqny information form whe whole webpage

company = soup.find_all('div', class_ = 'company-content-wrapper')
len(company)
```

```python
name = []
rating = []
reviews = []
ctype = []
hq = []
age = []
employees = []

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p', class_ = 'rating').text.strip())
    reviews.append(i.find('a', class_ = 'review-count').text.strip())
    ctype.append(i.find_all('p', class_ = 'infoEntity')[0].text.strip())
    #ctype.append(i.find('p', class_ = 'infoEntity').text.strip())   #->works
    hq.append(i.find_all('p', class_ = 'infoEntity')[1].text.strip())
    age.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip())
    try:
        employees.append(i.find_all('p', class_ = 'infoEntity')[3].text.strip())
    except:
        employees.append(np.nan)
    
```

```python
d = {'name':name, 'rating':rating, 'reviews':reviews, 'company-type':ctype, 'headquater':hq, 'age':age, 'no of employees':employees}

df = pd.DataFrame(d)
df
```

```python
final = pd.DataFrame()

#total pages in website =333
for j in range(1,334):
    
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text
    
    soup = BeautifulSoup(webpage, 'lxml')
    
    company = soup.find_all('div', class_ = 'company-content-wrapper')
    
    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    age = []
    employees = []

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p', class_ = 'rating').text.strip())
        reviews.append(i.find('a', class_ = 'review-count').text.strip())
        
        try:
            ctype.append(i.find_all('p', class_ = 'infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
            
        try:    
            hq.append(i.find_all('p', class_ = 'infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)

        try:

            age.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            age.append(np.nan)
        
        try:
            employees.append(i.find_all('p', class_ = 'infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)

    d = {'name':name, 'rating':rating, 'reviews':reviews, 'company-type':ctype, 'headquater':hq, 'age':age, 'no of employees':employees,}

    df = pd.DataFrame(d)
    final = final.append(df, ignore_index = True)
```
